```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=lg+washing+machine&sid=j9e%2Cabm%2C8qx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&as-pos=1&as-type=RECENT&suggestionId=lg+washing+machine%7CWashing+Machines&requestId=972f8ff1-b07e-4ffc-b9aa-202547e11628&as-searchtext=lg%20washing%20'
```

```python
r=requests.get(url)
```

```python
r
```

```python
print(r.text)
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

```python
soup
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
model=soup.findAll('div',{'class':'_4rR01T'})
```

```python
model
```

```python
print(product_name)
```
